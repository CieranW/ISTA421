Question 8
Source: https://stackoverflow.com/questions/14714181/conditional-logic-on-pandas-dataframe

```python
Q7["AgeGroup"] = Q7["Age"].apply(lambda x: 1 if x < 20 or x > 30 else 0)
Q8 = Q7.copy()

print(Q8)
```

Question 10

```python
def function_Q10(height, weight):
    return weight / (height**2)


Q10 = pd.DataFrame({"Height": height["value"], "Weight": weight["value"]})
Q10["BMI"] = function_Q10(Q10["Height"], Q10["Weight"])
print(Q10)
```

Question 11
Source: https://builtin.com/data-science/pandas-filter

```python
def function_Q11(df, targetGender):
    meanBMI = df.loc[df["Gender"] == targetGender, "BMI"].mean()
    return meanBMI


meanBMI = function_Q11(Q9, "A")
print(meanBMI)
```

Question 12
